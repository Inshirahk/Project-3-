```python
# Import dependencies
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Read the data into a Pandas DataFrame
od_data_df1 = pd.read_csv('data/Provisional_drug_overdose_death_counts_for_specific_drugs_20250128.csv')
od_data_df1.head()
```

```python
# Checking dataframes columns and datatypes
od_data_df1.info()
```

```python
# Removing the extra columns and rows with null values from the dataframe 
columns = ['footnote', 'time_period']

PDOD_specific_df = od_data_df1.drop(columns=columns)
PDOD_specific_df.dropna()
```

```python
# Changing data types of the columns 
PDOD_specific_df['data_as_of'] = pd.to_datetime(PDOD_specific_df['data_as_of'], format='%m/%d/%Y')

PDOD_specific_df['death_year'] = PDOD_specific_df['death_year'].astype(int)

PDOD_specific_df['death_month'] = PDOD_specific_df['death_month'].astype(int)

PDOD_specific_df['jurisdiction_occurrence'] = PDOD_specific_df['jurisdiction_occurrence'].astype(str)

PDOD_specific_df['drug_involved'] = PDOD_specific_df['drug_involved'].astype(str)

PDOD_specific_df['month_ending_date'] = pd.to_datetime(PDOD_specific_df['month_ending_date'], format='%m/%d/%Y')

PDOD_specific_df['drug_overdose_deaths'] = PDOD_specific_df['drug_overdose_deaths'].astype(float)

PDOD_specific_df
```

```python
# Checking dataframes columns and datatypes after using astype
PDOD_specific_df.info()
```

```python
# Saving the dataframe to a CSV
PDOD_specific_df.to_csv('data/PDOD_specific_cleaned.csv', index=False)
```

```python
# Read the data into a Pandas DataFrame
od_data_df2 = pd.read_csv('data/VSRR_Provisional_Drug_Overdose_Death_Counts_20250128.csv')
od_data_df2.head()
```

```python
# Checking dataframes columns and datatypes
od_data_df2.info()
```

```python
# Removing the extra columns and rows with null values from the dataframe 
columns = ['Footnote', 'Footnote Symbol','Predicted Value','Percent Pending Investigation','Percent Complete','Period']

VSRR_df = od_data_df2.drop(columns=columns)

VSRR_df.dropna()
```

```python
# Changing data types of the columns and adding the reigon column
VSRR_df['State'] = VSRR_df['State'].astype(str)

VSRR_df['State Name'] = VSRR_df['State Name'].astype(str)

VSRR_df['Year'] = VSRR_df['Year'].astype(int)

VSRR_df['Month'] = pd.to_datetime(VSRR_df['Month'], format='%B').dt.month

VSRR_df['Indicator'] = VSRR_df['Indicator'].astype(str)

VSRR_df["Region"] = 0
VSRR_df['Region'] = VSRR_df['Region'].astype(int)

VSRR_df
```

```python
# Creating the state reigon dictionary 
regions = {
    1: ["Connecticut", "Maine", "Massachusetts", "New Hampshire", "Rhode Island", "Vermont"],
    2: ["New Jersey", "New York", "Pennsylvania"],
    3: ["Illinois", "Indiana", "Michigan", "Ohio", "Wisconsin"],
    4: ["Iowa", "Kansas", "Minnesota", "Missouri", "Nebraska", "North Dakota", "South Dakota"],
    5: ["Delaware", "District of Columbia", "Florida", "Georgia", "Maryland", "North Carolina", "South Carolina", "Virginia",  "West Virginia"],
    6: ["Alabama", "Kentucky", "Mississippi", "Tennessee"],
    7: ["Arkansas", "Louisiana", "Oklahoma", "Texas"],
    8: ["Arizona", "Colorado", "Idaho", "Montana", "Nevada", "New Mexico", "Utah", "Wyoming"],
    9: ["Alaska", "California", 'Hawaii', "Oregon", "Washington"]
}
```

```python
# Updating the region column for each row using the dictionary
for index, row in VSRR_df.iterrows():
    for key, value in regions.items():
        if row['State Name'] in value:
            VSRR_df.at[index, 'Region'] = key
            break
```

```python
# Getting rid of all rows with data values of zero then saving the dataframe to a CSV 
VSRR_df = VSRR_df[VSRR_df['Data Value'] != 0]

VSRR_df.to_csv('data/VSRR_cleaned.csv', index=False)
```
