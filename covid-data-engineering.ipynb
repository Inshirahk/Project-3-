```python
# Import dependencies
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Read the data into a Pandas DataFrame
od_data_df1 = pd.read_csv('data/Provisional_drug_overdose_death_counts_for_specific_drugs_20250128.csv')
od_data_df1.head()
```

```python
#Removing the extra columns for the 
columns = ['footnote', 'time_period']

od_cols_df1 = od_data_df1.drop(columns=columns)
od_cols_df1.dropna()
```

```python
#change data types
od_cols_df1['data_as_of'] = pd.to_datetime(od_cols_df1['data_as_of'], format='%m/%d/%Y')

od_cols_df1['death_year'] = od_cols_df1['death_year'].astype(int)#date

od_cols_df1['death_month'] = od_cols_df1['death_month'].astype(int)#date

od_cols_df1['jurisdiction_occurrence'] = od_cols_df1['jurisdiction_occurrence'].astype(str)

od_cols_df1['drug_involved'] = od_cols_df1['drug_involved'].astype(str)

od_cols_df1['month_ending_date'] = pd.to_datetime(od_cols_df1['month_ending_date'], format='%m/%d/%Y')

od_cols_df1['drug_overdose_deaths'] = od_cols_df1['drug_overdose_deaths'].astype(float)

od_cols_df1
```
